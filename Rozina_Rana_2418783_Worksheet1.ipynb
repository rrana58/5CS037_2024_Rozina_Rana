Problem - 1: Array Creation:
Complete the following Tasks:
1. Initialize an empty array with size 2X2.
2. Initialize an all one array with size 4X2.
3. Return a new array of given shape and type, filled with fill value.{Hint: np.full}
4. Return a new array of zeros with same shape and type as a given array.{Hint: np.zeros like}
5. Return a new array of ones with same shape and type as a given array.{Hint: np.ones like}
6. For an existing list new_list = [1,2,3,4] convert to an numpy array.{Hint: np.array()}

```python
import numpy as np

empty_array = np.empty((2, 2))
print("Empty Array :")
print(empty_array)

ones_array = np.ones((4, 2))
print("\nAll Ones Array :")
print(ones_array)

full_array = np.full((3, 3), 7)
print("\nFull Array :")
print(full_array)

reference_array = np.array([[1, 2], [3, 4]])
zeros_array = np.zeros_like(reference_array)
print("\nZeros Array :")
print(zeros_array)

ones_array_like = np.ones_like(reference_array)
print("\nOnes Array :")
print(ones_array_like)

new_list = [1, 2, 3, 4]
numpy_array = np.array(new_list)
print("\nConverted NumPy Array:")
print(numpy_array)
```

Complete all the task using numpy:

{Hint:try np.concatenate() or np.vstack() functions.

```python
import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])


sum_array = x + y
print("Sum of x and y:")
print(sum_array)

difference_array = x - y
print("\nDifference of x and y:")
print(difference_array)

multiplication_result = x * 3
print("\nArray x multiplied by 3:")
print(multiplication_result)


squared_array = np.square(x)
print("\nSquare of each element of array x:")
print(squared_array)

dot_v_w = np.dot(v, w)
dot_x_v = np.dot(x, v)
dot_x_y = np.dot(x, y)

print("\nDot product of v and w:", dot_v_w)
print("Dot product of x and v:", dot_x_v)
print("Dot product of x and y:", dot_x_y)


concat_x_y_row = np.concatenate((x, y), axis=1)
print("\nConcatenate x and y along rows:")
print(concat_x_y_row)

v_w_column = np.vstack((v, w))
print("\nConcatenate v and w along columns:")
print(v_w_column)

try:
    concat_x_v = np.concatenate((x, v), axis=1)
    print("\nConcatenate x and v:")
    print(concat_x_v)
except Exception as e:
    print("\nError in concatenating x and v:")
    print(e)


v_reshaped = v.reshape(-1, 1)
concat_x_v = np.concatenate((x, v_reshaped), axis=1)
print("\nConcatenate reshaped x and v along rows:")
print(concat_x_v)
```

For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:
1. Prove A.A−1 = I.
2. Prove AB ̸= BA.
3. Prove (AB)


```python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
inverse=np.linalg.inv(A)
# print("Inverse of A: \n",inverse)
dotproduct=np.dot(A,inverse)
print(dotproduct)

c=np.dot(A,B)
print("AB: \n",c)
d=np.dot(B,A)
print("BA: \n",d)
print("Hence AB doesnt equal BA")

e=np.dot(A,B)
transpose1=np.transpose((e))
print("(AB)^t: \n",transpose1)

transpose2=np.transpose((A))
transpose3=np.transpose((B))
f=np.dot(transpose3,transpose2)
print("B^t.A^t: \n", f)
print("Hence prove (AB)^t = B^tA^t.")
```

```python
import time
import numpy as np

# Size for element-wise operations
size = 1000000

# Create two lists of size 1,000,000
list1 = [i for i in range(size)]
list2 = [i for i in range(size)]

# 1. Element-wise Addition using Python Lists
start_time = time.time()
result_list = [list1[i] + list2[i] for i in range(size)]
end_time = time.time()
print(f"Time taken for element-wise addition using Python lists: {end_time - start_time} seconds")

# Element-wise Addition using NumPy Arrays
array1 = np.arange(size)
array2 = np.arange(size)
start_time = time.time()
result_numpy = array1 + array2
end_time = time.time()
print(f"Time taken for element-wise addition using NumPy arrays: {end_time - start_time} seconds")

# 2. Element-wise Multiplication using Python Lists
start_time = time.time()
result_multiply_list = [list1[i] * list2[i] for i in range(size)]
end_time = time.time()
print(f"Time taken for element-wise multiplication using Python lists: {end_time - start_time} seconds")

# Element-wise Multiplication using NumPy Arrays
start_time = time.time()
result_multiply_numpy = array1 * array2
end_time = time.time()
print(f"Time taken for element-wise multiplication using NumPy arrays: {end_time - start_time} seconds")

# 3. Dot Product using Python Lists
start_time = time.time()
dot_product_list = sum(list1[i] * list2[i] for i in range(size))
end_time = time.time()
print(f"Time taken for dot product using Python lists: {end_time - start_time} seconds")

# Dot Product using NumPy Arrays
start_time = time.time()
dot_product_numpy = np.dot(array1, array2)
end_time = time.time()
print(f"Time taken for dot product using NumPy arrays: {end_time - start_time} seconds")

# 4. Matrix Multiplication using Python Lists
matrix1 = [[i + j for j in range(1000)] for i in range(1000)]
matrix2 = [[i - j for j in range(1000)] for i in range(1000)]

start_time = time.time()
result_matrix_list = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(1000)) for j in range(1000)] for i in range(1000)]
end_time = time.time()
print(f"Time taken for matrix multiplication using Python lists: {end_time - start_time} seconds")

# Matrix Multiplication using NumPy Arrays
matrix1_numpy = np.random.rand(1000, 1000)
matrix2_numpy = np.random.rand(1000, 1000)

start_time = time.time()
result_matrix_numpy = np.dot(matrix1_numpy, matrix2_numpy)
end_time = time.time()
print(f"Time taken for matrix multiplication using NumPy arrays: {end_time - start_time} seconds")
```
